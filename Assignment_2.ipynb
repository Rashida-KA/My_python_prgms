```python
df=pd.read_csv('data.csv')
df
```

```python
df.isna().sum()
```

```python
df['Calories'].fillna(df['Calories'].mean(), inplace=True)
df
```

```python
df["Date"].fillna("2020/12/22", inplace = True)
df
```

```python
df['Date']= pd.to_datetime(df['Date'])
df
```

```python
sn.relplot(data=df,x='Pulse',y='Calories')
font1={'family':'serif','color':'red','size':30}
plt.title("Visualization",fontdict=font1)
```

```python
sn.relplot(data=df,x='Duration',y='Calories',kind='line')
plt.grid(color='green',linestyle='--',linewidth=0.5)
```

```python
sn.relplot(data=df,x='Duration',y='Maxpulse',hue='Calories',color='r')
```

```python
sn.relplot(data=df,x='Calories',y='Date',style='Duration',kind='line',color='r')
```

```python
sn.relplot(data=df,x='Calories',y='Pulse',hue='Duration',kind='line')
```
