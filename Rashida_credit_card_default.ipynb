```python
# Import necessary libraries
import pandas as pd
import numpy as np
```

```python
# Load the credit_card_default dataset

df = pd.read_csv("credit_card_default.csv")
df
```

```python
df.columns
```

## Preprocessing

```python
df = df.rename(columns = lambda x: x.lower())
```

```python
df = df.rename(columns = {"default payment next month":"default"})
```

```python
df.duplicated().sum()
```

```python
df['sex'].value_counts()
```

## EDA 

```python
df["male"] = (df['sex'] == 1).astype(int)
df
```

```python
df['grad_school'] = (df['education']==1).astype(int)
df
```

```python
df["university"] = (df['education']==2).astype(int)
df
```

```python
df["married"] = (df['marriage']==1).astype(int)
df
```

```python
df.columns
```

```python
bill_amt_features = ["bill_amt" + str(i) for i in range(1,7)]
pay_amt_features = ["pay_amt" + str(i) for i in range(1,7)]
```

```python
binary_features = ['male', 'grad_school', 'university', 'married']
```

```python
pay_features = ["pay_" + str(i) for i in range(1,7)]
```

```python
 num_features = ["limit_bal", "age"] + bill_amt_features + pay_amt_features + pay_features
```

```python
# Split the dataset into features and target variable

x = df[num_features + binary_features]
y = df['default']
```

```python
# Split the dataset into training and test sets

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,
                                                    test_size=0.2, 
                                                    random_state = 134)
```

```python
sc_features = ["limit_bal", "age"] + bill_amt_features + pay_amt_features
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
x_train.loc[:,sc_features] = sc.fit_transform(x_train[sc_features])
```

```python
x_test.loc[:,sc_features] = sc.fit_transform(x_test[sc_features])
```

```python
import warnings
warnings.filterwarnings('ignore')
```

# KNN using Hyperparameter Tuning

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
```

```python
param_grid = {
    'n_neighbors': [3, 5, 7, 9],
    'weights': ['uniform', 'distance'],
    'p': [1, 2]
}
```

```python
from sklearn.model_selection import GridSearchCV
grid_knn = GridSearchCV(knn, param_grid, cv=5)
grid_knn.fit(x_train, y_train)
```

```python
print("Best hyperparameters:", grid_knn.best_params_)
```

```python
print("Best accuracy score:", grid_knn.best_score_)
```

```python
y_knn = grid_knn.best_estimator_.predict(x_test)
```

```python
# Calculate and print the accuracy score,precision,Recall,F1 score of the best model on the test data

from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score
print('Accuracy: ' , accuracy_score(y_test,y_knn)*100)
print('Precision: ', precision_score(y_test,y_knn)*100)
print('Recall: ', recall_score(y_test,y_knn)*100)
print('f1 score: ', f1_score(y_test,y_knn)*100)

```

# XGboost with hyperparameter tuning

```python
import xgboost as xgb
```

```python
# Define the parameter grid for the XGBoost classifier
param_grid = {
    'learning_rate': [0.01, 0.1, 1],
    'max_depth': [3, 5, 7],
    'n_estimators': [50, 100, 200]
}
```

```python
# Create an XGBoost classifier
xgb_classifier = xgb.XGBClassifier()
```

```python
# Create a grid search object with the parameter grid and cross-validation of 5
from sklearn.model_selection import GridSearchCV
grid_xgb = GridSearchCV(xgb_classifier, param_grid, cv=5, scoring='accuracy')
```

```python
# Fit the grid search object to the training data
grid_xgb.fit(x_train, y_train)
```

```python
# Print the best hyperparameters and the corresponding accuracy score
print("Best hyperparameters:", grid_xgb.best_params_)
print("Best accuracy score:", grid_xgb.best_score_)
```

```python
# Predict the test data using the best model found by the grid search object
y_pred_xgb = grid_xgb.best_estimator_.predict(x_test)
```

```python
y_pred_xgb
```

```python
print('Accuracy: ' , accuracy_score(y_test,y_pred_xgb)*100)
print('Precision: ', precision_score(y_test,y_pred_xgb)*100)
print('Recall: ', recall_score(y_test,y_pred_xgb)*100)
print('f1 score: ', f1_score(y_test,y_pred_xgb)*100)
```
